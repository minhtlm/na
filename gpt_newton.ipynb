```python
from sympy import *
```

```python
lap = 4

x, t = symbols('x t')
f = lambda x : x**3 - x**2 - 3

df = lambda x : f(t).diff().subs(t, x)
```

```python
plot(f(x), (x, 1, 4))
```

```python
plot(abs(f(x).diff(x, 2)), (x, 1, 4))
```

```python
M = 22.5
m = min( abs(df(1)), abs(df(4)))
m
```

```python
x0 = 4

for _ in range(lap):
    x = x0 - (f(x0) / df(x0))
    ss = (M / 2*m) * (x - x0)**2
    x0 = x
    print(N( x, 6), N(ss, 6))
```
