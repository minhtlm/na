```python
from sympy import *
```

```python
x = symbols('x')
g = lambda x : (x + 1) ** (1/4) 
```

```python
plot(g(x), (x, 1, 3))
```

```python
plot(abs(g(x).diff()), (x, 1, 3))
```

```python
q = 0.15

x0 = 1.5
n = 3

for _ in range(n):
    x = g(x0)
    ss = q / (1-q) * abs(x - x0)
    x0 = x
    print(x, ss)
```
